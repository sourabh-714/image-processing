## Smoothing Filters 

### 1. Averaging Filter 

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
 
img = cv2.imread('i1.jpg', 0) 
```

```python
plt.imshow(img)
plt.show()
```

```python
# rows and columns 
m, n = img.shape 
```

```python
# averaging filter developed of mask size 3*3
mask = np.ones([3,3], dtype = int) 
mask = mask / 9
```

```python
img_new = np.zeros([m, n]) 
  
for i in range(1, m-1): 
    for j in range(1, n-1): 
        temp = img[i-1, j-1]*mask[0, 0]+img[i-1, j]*mask[0, 1]+img[i-1, j + 1]*mask[0, 2]+img[i, j-1]*mask[1, 0]+ img[i, j]*mask[1, 1]+img[i, j + 1]*mask[1, 2]+img[i + 1, j-1]*mask[2, 0]+img[i + 1, j]*mask[2, 1]+img[i + 1, j + 1]*mask[2, 2] 
         
        img_new[i, j]= temp 
          
img_new = img_new.astype(np.uint8) 
```

```python
plt.imshow(img_new)
plt.show()
```

```python
### 2. Median Filter 
```

```python
# center pixel in grp of 3*3 area is replaced by the median
img_new1 = np.zeros([m, n]) 
```

```python
  
for i in range(1, m-1): 
    for j in range(1, n-1): 
        temp = [img[i-1, j-1], 
               img[i-1, j], 
               img[i-1, j + 1], 
               img[i, j-1], 
               img[i, j], 
               img[i, j + 1], 
               img[i + 1, j-1], 
               img[i + 1, j], 
               img[i + 1, j + 1]] 
          
        temp = sorted(temp) 
        img_new1[i, j]= temp[4] 
```

```python
img_new1 = img_new1.astype(np.uint8) 
```

```python
plt.imshow(img_new1)
plt.show()
```
