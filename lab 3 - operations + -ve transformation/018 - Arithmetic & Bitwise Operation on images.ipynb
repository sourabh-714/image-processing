### Arithmetic Operations 

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
Img1 = cv2.imread('img1.jpg')
Img2 = cv2.imread('img2.jpg')
```

```python
Img1.shape
```

```python
plt.imshow(Img1)
```

```python
plt.imshow(Img2)
```

```python
Img2.shape
```

```python
add = cv2.add(Img1,Img2)
```

```python
plt.imshow(add)
```

```python
sub = cv2.subtract(Img1,Img2)
```

```python
plt.imshow(sub)
```

```python
mul = cv2.multiply(Img1,Img2)
```

```python
plt.imshow(mul)
```

```python
div = cv2.divide(Img1,Img2)
```

```python
plt.imshow(div)
```

### Bitwise Operations 

```python
add = cv2.bitwise_and(Img1,Img2)
```

```python
plt.imshow(add)
```

```python
or1 = cv2.bitwise_or(Img1,Img2)
```

```python
plt.imshow(or1)
```

```python
not1 = cv2.bitwise_not(Img1)
```

```python
plt.imshow(not1)
```

```python
xor = cv2.bitwise_xor(Img1,Img2)
```

```python
plt.imshow(xor)
```
