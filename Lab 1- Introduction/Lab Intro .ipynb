```python
import matplotlib.pyplot as plt
```

```python
from PIL import Image
```

```python
img = Image.open('minion.jpg')
```

```python
img
```

```python
type(img)
```

```python
img = img.resize((300,300))
img
```

```python
arr = np.asarray(img)
arr.shape
```

```python
arr1 = arr.copy()
```

```python
arr1[:,:,1] = 0 # zeroed the effect of colour Green
arr1[:,:,2] = 0 # Zeroed the effect of colour Blue
```

```python
plt.imshow(arr1)
```

```python
arr2 = arr.copy()
```

```python
arr2[:,:,0] = 0 # effect of Red Zeroed
plt.imshow(arr2) # image contains a mix of Green and Blue
```

```python
arr3 = arr.copy()
```

```python
arr3[:,:,0] = 0
arr3[:,:,1] = 0
arr3[:,:,2] = 0
# all 3 channels are made zero and hence a black image
```

```python
plt.imshow(arr3)
```
