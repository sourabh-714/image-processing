### 1. Contrast Stretching 

```python
import cv2
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('i2.jfif')

plt.imshow(img)
plt.show()
plt.hist(img.ravel(),256,[0,256])
plt.show()
```

```python
def pixelVal(pix, min, max): 
    return (((pix - min) / (max - min)) * 255 )
```

```python
max = img.max()
```

```python
min = img.min()
```

```python
# Vectorize the function to apply it to each value in the Numpy array. 
pixelVal_vec = np.vectorize(pixelVal) 
  
# Apply contrast stretching. 
contrast_stretched = pixelVal_vec(img, min, max) 
  
plt.imshow(contrast_stretched)
plt.show()

plt.hist(contrast_stretched.ravel(),256,[0,256])
plt.show()
```

### 2. Thresholding

```python
img = cv2.imread('i2.jfif')

plt.imshow(img)
plt.show()
```

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
```

```python
ret, thresh1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY) 
plt.imshow(thresh1)
```

### 3. Gray Level Slicing 

```python
# Find width and height of image
row, column, _ = img.shape
```

```python
# Create an zeros array to store the sliced image
img4 = np.zeros((row,column),dtype = 'uint8')
```

```python
# Specify the min and max range
min_range = 10
max_range = 60
 
# Loop over the input image and if pixel value lies in desired range set it to 255 otherwise set it to 0.
for i in range(row):
    for j in range(column):
        if img2[i,j]>min_range and img2[i,j]<max_range:
            img4[i,j] = 255
        else:
            img4[i,j] = 0
```

```python
plt.imshow(img4) 
plt.show()
```

### 4. Histogram Equalisation 

```python
img = cv2.imread('i2.jfif',0)
equ = cv2.equalizeHist(img)
res = np.hstack((img,equ)) #stacking images side-by-side
plt.imshow(res)
plt.show()
```

### - Using the formula 

```python
img = cv2.imread('img1.jpg')

hist,bins = np.histogram(img.flatten(),256,[0,256])

cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()

plt.plot(cdf_normalized, color = 'b')
plt.hist(img.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

```python
cdf_m = np.ma.masked_equal(cdf,0)
cdf_m = (cdf_m - cdf_m.min())*255/(cdf_m.max()-cdf_m.min())
cdf = np.ma.filled(cdf_m,0).astype('uint8')
```

```python
img2 = cdf[img]
plt.plot(cdf_m, color = 'b')
plt.hist(img2.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

```python
res = np.hstack((img,img2))
plt.imshow(res)
plt.show()
```
