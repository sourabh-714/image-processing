```python
import cv2
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('i1.jpg',0)
plt.imshow(img)
```

### 1. Filter 2D

```python
kernel = np.ones((5,5), np.float32)/20
dst = cv2.filter2D(img,-1,kernel)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

### 2. Blur 

```python
dst = cv2.blur(img,(50,50))

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

### 3. Median Blur  

```python
dst = cv2.medianBlur(img,25)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

### 4. Box Filter

```python

dst = cv2.boxFilter(img,-1,(50,50), normalize= True)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```
